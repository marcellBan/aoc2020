```python
groups = []
with open('day6.in') as f:
    group = set()
    for line in f:
        if line == '\n':
            groups.append(group)
            group = set()
        else:
            for c in line.strip():
                group.add(c)
    if group:
        groups.append(group)
print(groups)
```

```python
print(f'Part 1 answer: {sum(len(x) for x in groups)}')
```

```python
groups = []
with open('day6.in') as f:
    group = set()
    allq = ''
    cnt = 0
    for line in f:
        if line == '\n':
            ng = set()
            for c in group:
                if allq.count(c) == cnt:
                    ng.add(c)
            groups.append(ng)
            group = set()
            allq = ''
            cnt = 0
        else:
            allq += line.strip()
            cnt += 1
            for c in line.strip():
                group.add(c)
    if group:
        ng = set()
        for c in group:
            if allq.count(c) == cnt:
                ng.add(c)
        groups.append(ng)
print(groups)
```

```python
print(f'Part 2 answer: {sum(len(x) for x in groups)}')
```
