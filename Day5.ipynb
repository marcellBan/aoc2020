```python
with open('day5.in') as f:
    lines = [(x[:7],x[7:-1]) for x in f]
    
trn = str.maketrans({'F':'0', 'B':'1', 'R':'1', 'L': '0'})
    
translated = [(x[0].translate(trn), x[1].translate(trn)) for x in lines]
print(translated)
```

```python
all_seats = sorted([int(x[0],base=2)*8+int(x[1],base=2) for x in translated])
print(f'Part 1 answer: {all_seats[-1]}')
```
