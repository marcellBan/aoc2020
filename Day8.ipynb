```python
with open('day8.in') as f:
    program = [tuple(x.split()) for x in f]
print(program)
```

```python
def run(program):
    acc = 0
    pc = 0
    seen = set()
    while pc < len(program):
        if pc in seen:
            return False, acc
        seen.add(pc)
        if program[pc][0] == 'nop':
            pc += 1
        elif program[pc][0] == 'acc':
            acc += int(program[pc][1])
            pc += 1
        elif program[pc][0] == 'jmp':
            pc += int(program[pc][1])
    return True, acc

print(f'Part 1 answer: {run(program)[1]}')
```

```python
to_check = [x[0] for x in filter(lambda i: i[1][0]=='nop' or i[1][0]=='jmp',enumerate(program))]
print(to_check)
```

```python
switch = {'nop': 'jmp', 'jmp': 'nop'}
for ch in to_check:
    mprog = list(program)
    mprog[ch] = (switch[mprog[ch][0]], mprog[ch][1])
    smooth, acc = run(mprog)
    if smooth:
        print(f'Part 2 answer: {acc}')
        break
    
```
