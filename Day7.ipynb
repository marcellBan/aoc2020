```python
with open('day7.in') as f:
    lines = [x.split() for x in f]
print(lines)
```

```python
matrix = {}
for line in lines:
    k = tuple(line[:2])
    matrix[k] = []
    line = line[4:]
    while line:
        matrix[k].append(tuple(line[1:3]))
        line = line[4:]
        
print(matrix)
```

```python
to_check = [('shiny','gold')]
seen = set()
while to_check:
    curr = to_check[0]
    seen.add(curr)
    for k,v in matrix.items():
        if k not in seen and curr in v:
            to_check.append(k)
    del to_check[0]
print(f'Part 1 answer: {len(seen)-1}')
```

```python
matrix = {}
for line in lines:
    k = tuple([1,line[0],line[1]])
    matrix[k] = []
    line = line[4:]
    while line:
        if line[0] != 'no':
            matrix[k].append(tuple(line[:3]))
        line = line[4:]
```

```python
to_sum =[(1,'shiny','gold')]
summa=0
while to_sum:
    curr = to_sum[0]
    v = matrix[tuple([1,curr[1],curr[2]])]
    for b in v:
        summa += curr[0]*int(b[0])
        to_sum.append(tuple([curr[0]*int(b[0]),b[1],b[2]]))
    del to_sum[0]

print(f'Part 2 answer: {summa}')
```
